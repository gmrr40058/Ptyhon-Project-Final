```python
x = np.linspace(0, 10, 100)
x
```

```python
s = np.sin(x)
```

```python
c = np.cos(x)
```

```python
fig = plt.figure()
plt.plot(x, s)
plt.plot(x, c, ls ='--')
plt.xlabel("X value")
plt.ylabel('Blue for sin and orange for cos')
plt.show()
```

```python
fig.canvas.get_supported_filetypes()
```

```python
fig.savefig("Sin and Cos curve.jpg")
```

```python
fig = plt.figure()
plt.plot(x, s, marker  = '.')
plt.plot(x, c, ls ='--')
plt.xlabel("X value")
plt.ylabel('Blue for sin and orange for cos')
plt.grid()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, s, marker  = '.')
plt.title("Sin curve")
plt.xlabel("X value")
plt.ylabel('Sin X')
plt.grid()

plt.subplot(2,1,2)
plt.plot(x, c, ls ='-')
plt.title("Cos curve")
plt.xlabel("X value")
plt.ylabel('Cos X')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, s, marker  = '.')
plt.title("Sin curve")
plt.xlabel("X value")
plt.ylabel('Sin X')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x, c, ls ='-')
plt.title("Cos curve")
plt.xlabel("X value")
plt.ylabel('Cos X')
plt.grid()
plt.show()
```
