```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.__version__
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.lineplot(x = 'sepal_length', y = 'sepal_width', data = iris)
plt.title("Iris plot")
plt.show()
```

```python
sns.stripplot(x = 'species', y = 'sepal_length', data = iris)
plt.title("Iris plot")
plt.show()
```

```python
plt.subplot(321)
sns.stripplot(x='day', y='total_bill', data=tips)

plt.subplot(322)
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)

plt.subplot(323)
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker')

plt.subplot(324)
sns.stripplot(x='day', y='total_bill', data=tips, dodge=True)

plt.subplot(325)
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='smoker', dodge=True)
```

```python
sns.stripplot(x='species', y='sepal_length', data=iris)
plt.title("Stripplot based on 'iris' dataset")
```

```python
sns.stripplot(x='day', y='total_bill', data=tips,
 jitter=True,
 hue='smoker',
 dodge=True
 )
plt.title("Stripplot based on 'tips' dataset");
```

```python
sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title("Swarmplot based on 'iris' dataset");
```

```python
sns.barplot(x="sex", y="tip", data=tips)
plt.title("Barplot based on 'tips' dataset");
```

```python
sns.barplot(x="total_bill", y="sex", data=tips,
 palette="plasma")
plt.title("Another Barplot based on 'tips' dataset");
```

```python
sns.barplot(x="total_bill", y="sex", data=tips)
plt.title("Another Barplot based on 'tips' dataset");
```

```python
sns.barplot(x="sex", y="tip", data=tips)
plt.title("Barplot based on 'tips' dataset");
```

```python
sns.countplot(x="sex", data=tips,
 palette="rainbow")
plt.title("Countplot based on 'tips' dataset");
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
 
# Removing the spines
sns.despine() # by default, top=True, left=True
```

```python
flights = sns.load_dataset('flights')
plt.figure(figsize = (8, 5))
sns.lineplot(x="month", y="passengers", data=flights)
plt.title("Changing figure size using 'figure()' method of matplotlib");
```
